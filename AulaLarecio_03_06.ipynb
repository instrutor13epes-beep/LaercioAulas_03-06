# Entrada de texto com input():

Exemplo completo de código Python:

Com o comando input temos a base para fazer programas de cadastro, tipo quando vocês fazem um login em uma plataforma ou preenche um cadastro em sites:

Experimente o código abaixo:



```
nome = input("Qual é o seu nome? ")
cidade = input("Em que cidade você mora?")
print("Olá" , nome , "! Você mora em", cidade)
```
Quando criamos uma caixa de texto, guardamos o que o úsuario escreve em uma **Variável**:

**Variável** é uma forma de guardar(armazenar) uma informação ligada a um nome e com isso podemos reeutilizar no código:

No nosso código as variáveis são usadas aqui:
```
print("Olá" , nome , "! Você mora em", cidade)
```



````python
nome = input("Qual é o seu nome? ")
cidade = input("Em que cidade você mora? ")
print("Olá" , nome , "! Você mora em", cidade)
````
